```python
import torch
import torchvision
from torchvision import transforms
```

```python
trans = transforms.Compose(
    [
        transforms.ToTensor()
    ]
)
mninst_train = torchvision.datasets.FashionMNIST(
    root='FashionMINIST',
    train=True,
    download=True,
    transform=trans
    )
mninst_text = torchvision.datasets. FashionMNIST(
    root="FashionMINIST",
    train=False,
    download=True,
    transform=trans
)


len(mninst_train),len(mninst_text)
```

```python
batch_size = 256 
train_dataloader = torch.utils.data.DataLoader(
    dataset=mninst_train,
    batch_size=batch_size,
    shuffle=True,
    drop_last=True,
    num_workers=4
)
text_dataloader = torch.utils.data.DataLoader(
    dataset=mninst_text,
    batch_size=batch_size,
    shuffle=True,
    drop_last=True,
    num_workers=4
)
```

```python
num_inputs = 28*28 
num_out1 = 256
w1 = torch.normal(0,0.01,size=(num_inputs,num_out1),requires_grad=True)
b1 = torch.zeros(num_out1,requires_grad=True)
num_out2 = 10
w2 = torch.normal(0,0.01,size=(num_out1,num_out2),requires_grad=True)
b2 = torch.zeros(num_out2,requires_grad=True)
```

```python

def softmax(X):
    exp_x = torch.exp(X)
    p = exp_x.sum(1,keepdim=True)
    return exp_x / p
def net(w,b,X):
    return torch.matmul(X.reshape((256,-1)),w) + b

def lossfunction(y_hat , y):

    return -torch.log(y_hat[range(y_hat.shape[0]),y])
```

```python
def relu(x):
    a = torch.zeros_like(x)
    return torch.max(x,a)
def sgd(paramm,lr):
    with torch.no_grad():
       for param in paramm:
            param -= lr*param.grad / 256
            param.grad.zero_()
def accuracy(y_hat,y):
    predictIndex = y_hat.argmax(axis=1)
    result = (predictIndex == y).sum()
    return result / y_hat.shape[0]
for _ in range(10):
    for X,y in train_dataloader:
        y_hat = net(w1,b1,X)
        y_hat = relu(y_hat)
        y_hat = net(w2,b2,y_hat)
        y_hat = softmax(y_hat)
        
        acc = accuracy(y_hat,y)

        loss = lossfunction(y_hat, y )

        loss.sum().backward()

        sgd([w1,b1,w2,b2],lr=0.01)
        
    print(loss.sum()/256)
    print(acc*100)
```

```python
from torch.nn import(
    Sequential,
    Flatten,
    CrossEntropyLoss,
    Linear,
    ReLU
)
from torch.optim import SGD
net = Sequential(
    Flatten(),
    Linear(784,256),
    ReLU(),
    Linear(256,10)

)
loss_ = CrossEntropyLoss()
optimer = SGD(net.parameters(),lr=0.01)
for _ in range(10):
    for X , y in train_dataloader:
        y_hat = net(X)
        acc = accuracy(y_hat,y)
        loss = loss_(y_hat , y)
        optimer.zero_grad()
        loss.backward()
        optimer.step()
    print(loss)
    print(acc)
```
