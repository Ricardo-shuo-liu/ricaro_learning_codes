```python
import numpy as np
def chose(x):
    result = np.random.uniform(low=-1,high=1,size=(x,2))
    return result

result = chose(1)
result
```

```python
def get_pi_with_calculating(X):
    result = chose(X)
    pi = ((result[:,0]**2 + result[:,1]**2)<=1).sum()*4/X
    return pi
pi = get_pi_with_calculating(1000000)
pi
```

```python
def get_area(X):
    def choose(X):
        result = np.random.uniform(low=0,high=2,size=(X,2))
        return result
    examples = choose(X)
    r = (examples[:,0]**2 + examples[:,1]**2)>4
    b = ((examples[:,0]-1)**2 + (examples[:,1]-1)**2)<=1
    s = r & b
    return s.sum()*4/X
are = get_area(1000)
are
```
