```python
def uniform(n):
    import random
    d = random.randint(0, n-1)
    return d 

def fisher_yates(input_list):
    lst = input_list.copy()
    a = []
    for i in range(len(lst)):
        index = uniform(len(lst) - i)
        a.append(lst[index])
        for x in range(index, len(lst)-i-1):
            lst[x] = lst[x+1]
    return a
```

```python
def fiser_yates_improve(input_list):
    lst = input_list.copy()
    i = 0
    for i in range(len(lst)-1,0,-1):
        index = uniform(i)
        lst[i],lst[index] = lst[index],lst[i]
        i+=1
    return lst
lst = fiser_yates_improve(['a','b','c','d'])
lst
```
