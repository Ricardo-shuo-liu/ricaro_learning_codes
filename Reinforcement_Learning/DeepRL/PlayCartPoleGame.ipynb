```python
import gym
env = gym.make('CartPole-v1',render_mode="human")
state = env.reset()

for _ in range(100):
    # env.render()
    print(state)
    action = env.action_space.sample()
    state, reward, terminated, truncated, info = env.step(action)
    done = terminated or truncated
    print(type(action))
    if done:
        print("Finishe")
        break
    

env.close()
```
